```python
#샘플 데이터 url
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

df = pd.read_csv(url)
df.head() #상위 5개의 행 보기
```

```python
# @title Age

from matplotlib import pyplot as plt
df['Age'].plot(kind='line', figsize=(8, 4), title='Age')
plt.gca().spines[['top', 'right']].set_visible(False)
```

# **😊😊데이터를 선택하고 필터링 분석**

```python
#하나의 열 선택
df['Age'].head()

#여러개의 열 선택
df[['Name', 'Age']].head()

#30세이상의 승객
df[df['Age'] >= 30].head()

#30세이상이면서 남성의 승객
df[(df['Age'] >= 30) & (df['Sex'] == 'male')].head()



```

# 😊😊기초통계 및 정렬

```python
#평균나이
df['Age'].mean()
```

```python
#생존자수 카운트
df['Survived'].value_counts()
```

```python
#나이를 내림차순 기준으로 정렬
df.sort_values(by='Age', ascending=False).head()
```

# 그룹별 분석😊😊

```python
#groupby()
#성별에 따른 생존률 평균
df.groupby('Sex')['Survived'].mean()
```

```python
#선실등급(Pclass)에 따른 평균나이
df.groupby('Pclass')['Age'].mean()
```

# 간단한 시각화 😊😊

```python
#성별에 따른 생존률 시각화
df.groupby('Sex')['Survived'].mean().plot(kind='bar', title='Survival Rate by Sex')
```

```python
#나이분포 히스토그램

df['Age'].hist(bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

자주 사용하는 Padas 함수 정리
1. 데이터 불러오기 --> pd.read_csv()
2. 열 선택 --> df['column']
3. 행 필터링 --> df[df['Age'] > 30]
4. 정렬 --> df.sort_value(by = 'Age')
5. 그룹화 --> df.groupby('Sex')
6. 결측치 처리 --> df.dropna(), df.fillna(value)
